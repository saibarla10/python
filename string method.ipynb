# captalize() -  it is used to change the first letter of a string to uppercase and make all other letters lowercase. It is especially useful when we want to ensure that text follows title-like capitalization, where only the first letter is capitalized.

```python
# examples
```

```python
s = "hello WORLD"
print(s.capitalize())
```

```python
word = "python Class"
print(word.capitalize())
```

```python
b = "arrow bow"
print(b.capitalize())
```

# casefold() - it is used to convert string to lowercase. It is similar to the Python lower() string method, but the case removes all the case distinctions present in a string.

```python
# examples
```

```python
string = "GEEKSFORGEEKS"

print("lowercase string: ", string.casefold())
```

```python
word = "HYDERABADCITY"
print(word.casefold())
```

```python
f = "TRAFFICCHALLAN"
print(f.casefold())
```

# center() - it is a simple and efficient way to center-align strings within a given width. By default, it uses spaces to fill the extra space but can also be customized to use any character we want.

```python
# examples
```

```python

s1 = "hello"
s2 = s1.center(20)
print(s2)
```

```python
a = "sai"
b = a.center(18)
print(b)
```
